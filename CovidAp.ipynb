```
import pandas as pd

dataset = pd.read_csv('covidap.csv',',' ,error_bad_lines=False)

dataset.head(10)
```

```
dataset['date'] = pd.to_datetime(dataset['date'])

dataset['date'] = dataset['date'].dt.strftime('%m/%d/%Y')
```

**Evolução dos Casos AP** *italicized text*

```
import matplotlib.pyplot as plt

estados = dataset[dataset['place_type'] == 'state']

estados = estados.sort_values(by=['date'], ascending=True)

plt.figure(figsize=(30,10))
plt.plot(estados.date ,estados.confirmed, label='Casos')
plt.plot(estados.date ,estados.deaths, label='Mortes')

plt.yscale('linear', linthreshy=10)

plt.ylabel('Número')
plt.xlabel('Data')
plt.title('Número de Casos AP')
plt.grid(True)
```

```
import matplotlib.pyplot as plt
cidades = dataset[dataset['place_type'] == 'city']
cidades = dataset[dataset['city'] == 'Macapá']

cidades = cidades.sort_values(by=['date'], ascending=True)

plt.figure(figsize=(30,10))
plt.plot(cidades.date ,cidades.confirmed, label='Casos Confirmados')
plt.plot(cidades.date ,cidades.deaths, label='Mortes')
plt.ylabel('Número')
plt.title('Número de Casos Macapá')

plt.grid(True)
plt.show()
```
